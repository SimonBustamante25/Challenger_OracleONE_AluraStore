# Conclusión y recomendación:


Basándonos en el análisis de los datos, especialmente en los ingresos totales y la evolución de los ingresos diarios, se recomienda vender la Tienda 4. A pesar de tener un costo de envío promedio ligeramente menor y una calificación promedio similar a las otras, sus ingresos son los más bajos en comparación con las otras tres tiendas. Si bien otras métricas como las ventas por categoría y los productos más/menos vendidos pueden ofrecer información detallada, la métrica de ingresos es la más relevante para determinar el rendimiento financiero general. #1. Análisis de facturación



Ingresos por tienda

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
ingresos = {}
contador1 = 1
for i in tiendas:
  ingreso = i["Precio"].sum()
  ingresos[f"Tienda {contador1}"] = int(ingreso)
  contador1 += 1

for tiendaa, valor in sorted(ingresos.items(), key=lambda x: x[1], reverse=True):
    print(f"{tiendaa}: ${valor:,.0f}")


# Ordenar los datos de mayor a menor
ingresos_ordenados = dict(sorted(ingresos.items(), key=lambda x: x[1], reverse=True))

# Crear el gráfico
plt.figure(figsize=(8, 5))
plt.bar(ingresos_ordenados.keys(), ingresos_ordenados.values(), color=['#4B8BBE', '#306998', '#FFE873', '#FFD43B'])

# Agregar etiquetas y título
plt.title("Comparación de ingresos por tiendaa")
plt.xlabel("Tienda")
plt.ylabel("Ingresos totales ($)")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Agregar valores encima de cada barra
for i, v in enumerate(ingresos_ordenados.values()):
    plt.text(i, v + (max(ingresos_ordenados.values()) * 0.01), f"${v:,.0f}", ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# Lista de dataframes y títulos
tiendas = [tienda, tienda2, tienda3, tienda4]
titulos = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# Crear la figura y los ejes
fig, axes = plt.subplots(2, 2, figsize=(14, 8))
axes = axes.flatten()

# Recorrer cada tienda y graficar
for ax, df, titulo in zip(axes, tiendas, titulos):
    # Agrupar por fecha y sumar ingresos
    df_ventas = df[["Fecha de Compra", "Precio"]].copy()
    df_ventas = df_ventas.groupby("Fecha de Compra")["Precio"].sum().reset_index()

    # Convertir fechas y filtrar años 2022–2023
    df_ventas["Fecha de Compra"] = pd.to_datetime(df_ventas["Fecha de Compra"], format='%d/%m/%Y', errors='coerce')
    df_ventas = df_ventas.dropna(subset=["Fecha de Compra"])
    df_ventas = df_ventas.sort_values("Fecha de Compra")
    df_ventas = df_ventas[(df_ventas["Fecha de Compra"].dt.year >= 2022) &
                          (df_ventas["Fecha de Compra"].dt.year <= 2023)]

    # Calcular el promedio de ingresos
    promedio = df_ventas["Precio"].mean()

    # Graficar la línea de ingresos diarios
    ax.plot(df_ventas["Fecha de Compra"], df_ventas["Precio"], color="royalblue", label="Ingresos diarios")

    # Añadir línea horizontal del promedio
    ax.axhline(promedio, color="red", linestyle="--", linewidth=1.5, label=f"Promedio: ${promedio:,.0f}")

    # Personalización
    ax.set_title(f"Ingresos por fecha ({titulo}) - 2022 y 2023")
    ax.set_xlabel("Fecha de Compra")
    ax.set_ylabel("Ingresos")
    ax.tick_params(axis='x', rotation=45)
    ax.grid(True, linestyle='--', alpha=0.5)
    ax.legend(fontsize=8)

# Ajustar espaciado
plt.tight_layout()
plt.show()
```
